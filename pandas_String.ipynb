#date_range function

```python
#genereate daily date ina given range
pd.date_range(start='2023/1/1',end='2023/2/1',freq='2D')
```

```python
#genreate bussiness days in a given range exclude satursday and sunday use B
pd.date_range(start='2023/1/1',end='2023/2/1',freq='B')
```

```python
#W --> one week perday (sunday)
pd.date_range(start='2023/1/1',end='2023/2/1',freq='W')
```

```python
# week ka ki bhi din
pd.date_range(start='2023/1/1',end='2023/2/1',freq='W-FRI')
```

```python
#H --> hourly data(factor)
pd.date_range(start='2023/1/1',end='2023/2/1',freq='H')
```

```python
#M--> month end
pd.date_range(start='2023/1/1',end='2023/2/1',freq='M')
```

```python
#M --> month startng
pd.date_range(start='2023/1/1',end='2023/2/1',freq='MS')
```

```python
#Y--> year end
pd.date_range(start='2023-01-01', end='2023-02-01', freq='YE')
```

```python
#periods
pd.date_range(start='2023/1/5',periods=25,freq='M')
```

to_datetime functions

converts an existing objects to pandas timestamp/datetimeindex object

```python
#simple series example
s=pd.Series(['2023/1/1','2026/2/2','2025/3/3'])
print(pd.to_datetime(s).dt.day_name())
```

```python
print(pd.to_datetime(s).dt.month_name())
```

```python
print(pd.to_datetime(s).dt.month)
```

```python
print(pd.to_datetime(s).dt.date)
```

```python
#with error agar error
#corec --> unvalid values ko ignore kr deta hai
s=pd.Series(['2023/1/1','2021/2/2'])
pd.to_datetime(s,errors='coerce')
```
