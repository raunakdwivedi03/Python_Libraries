#Broadcasting

The term broadcasting describe how numpy treats arrays with different shapes during airthemetic operations. the smaller array is broadcast acrros the larger array so that they have complicated shapes.


Broadcasting NumPy ka feature hai jisme chhota array automatically bada array ke size ke according adjust ho jata hai, taaki operation possible ho sake.

```python
##diff shape
a=np.arange(6).reshape(2,3)
b=np.arange(3).reshape(1,3)

print(a)
print(b)


print(a+b)
```

#Rules of broadcasting

ðŸ”¹ Rule 1: Dimensions equal hone chahiye

Agar dono arrays ki dimension same hai, to broadcasting possible hai.


ðŸ”¹ Rule 2: Kisi ek dimension ka size 1 hona chahiye

Agar dimension same nahi hai, to kisi ek array ka size 1 hona chahiye.

ðŸ”¹ Rule 3: Missing dimension ko 1 maana jata hai

Agar kisi array me kam dimensions hain, to left side se 1 add kiya jata hai.

```python
##More example
## !d hai toh usko 2D banana hai
a=np.arange(12).reshape(4,3)
b=np.arange(3)

print(a)
print(b)

print(a+b)
```

```python
a=np.arange(12).reshape(4,3)
b=np.arange(4).reshape(4,1)

print(a)
print(b)

print(a+b)
```

```python
##mean squared error --> losss fucntion in linear regression
# Mean Squared Error ek loss / error function hai jo batata hai ki predicted values aur
# actual values ke beech average squared difference kitna hai.


actual=np.random.randint(1,50,20)
predicted=np.random.randint(1,50,20)
```

```python
#Binary Cross Entropy ek loss function hai jo binary classification problems me use hota hai
# (jaha output sirf 0 ya 1 hota hai).
```

```python
## binary cross entropy is for logistic regression
import numpy as np

y = np.array([1, 0, 1, 0])
y_pred = np.array([0.9, 0.2, 0.8, 0.1])

bce = -np.mean(
    y * np.log(y_pred) + (1 - y) * np.log(1 - y_pred)
)

print(bce)
```
