```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv('/content/imdb-top-1000.csv')
```

```python
movies.head(1)
```

```python
genres=movies.groupby('Genre')
genres
```

#GroupBy fucntion

Pandas me groupby() function ka use hum data ko kisi column (ya columns) ke basis par groups me todne ke liye karte hain, aur phir har group par calculation / analysis karte hain

groupBy categorical data ke uppar use hota hai

```python
##Applying builtin aggregation function on groupBy objects
genres.sum()
```

```python
genres.min()
```

```python
std  = float(genres.std())
```

```python
#find the top 3 earning genres total earning
movies.groupby('Genre').sum()['Gross'].sort_values(ascending=False).head(3)
```

```python
#method 2 to find
movies.groupby('Genre')['Gross'].sum().sort_values(ascending=False).head(3)
```

```python
# find the direcctor who is most popular
movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(1)
```

```python
#find the highes rated movies of all genres
movies.groupby('Genre')['IMDB_Rating'].max()
```

```python
#find the number of movies done by each actor
movies['Star1'].value_counts()
```

```python
#method 2 using Groupby
movies.groupby('Star1')['Series_Title'].count().sort_values(ascending=False)
```

```python
#groupby attributes and method
#find the total numnbers of groups_>len

len(movies.groupby('Genre'))
```
