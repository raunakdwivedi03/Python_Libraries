#Array functions

```python
a1=np.random.random((3,3))
print(a1)
a1=np.round(a1*100)
print(a1)
```

```python
#max/min/sum/prod
#0--> col  and 1--> row
print(np.max(a1))
print(np.min(a1,axis=1))
print(np.prod(a1,axis=0))
print(np.sum(a1))
print(np.sum(a1,axis=0))
```

```python
##median/mean/std/var --> statistical operations
print(np.median(a1))
print(np.mean(a1))
print(np.std(a1))
print(np.var(a1))

print(np.median(a2))
print(np.mean(a2))
print(np.std(a2))
print(np.var(a2))

print(np.median(a2,axis=0))
print(np.mean(a2,axis=0))
print(np.std(a2,axis=0))
print(np.var(a2,axis=0))
```

```python
##trignometric operations
print(np.sin(a1))

print(np.cos(a1))

print(np.tan(a1))
```

```python
##dot product --> if (3,4) and (4,3) if in this ordered pair the i==j then it is call the dot product and the resultant is (3,3)
a2=np.arange(12).reshape(3,4)
a3=np.arange(12,24).reshape(4,3)
np.dot(a2,a3)
```

```python
##log and exponent
print(np.log(a1))
print(np.exp(a1))
```

```python
##round / floor / and ciel
print(np.random.random((2,3))*100)
print(np.round(np.random.random((2,3))*100)) ## o/p round off krke dega
```

```python
np.floor(np.random.random((2,3))*100) ## floor peeche vaale numbers ko deta hai
```
