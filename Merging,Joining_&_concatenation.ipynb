#Merge

```python
## inner join
regs
```

```python
students
```

```python
regs
```

```python
# inner join
students.merge(regs,how='inner',on='student_id')
```

```python
#left join
course.merge(regs,how='left',on='course_id')
```

```python
# right join
temp_df=pd.DataFrame({
    'student_id':[26,27,28],
     'name':['raunak','rahul','shyaam'],
    'partner':[28,26,17]

})

students=pd.concat([students,temp_df],ignore_index=True)
```

```python
students.tail()
```

```python
##right join
students.merge(regs,how='right',on='student_id')
```

```python
#outer join
students.merge(regs,how='outer',on='student_id').tail(10)
```

```python
## find total revenue generated
total_revenue_genrated=regs.merge(course,how='inner',on='course_id')['price'].sum()
total_revenue_genrated
```

```python
# find month by month revenue
temp_df=pd.concat([nov,dec],keys=['nov','dec']).reset_index()
temp_df.merge(course,on='course_id').groupby('level_0')['price'].sum()
```

```python
# print the registration table
#cols -> name -> course -> price
regs.merge(students,on='student_id').merge(course,on='course_id')[['name','course_name','price']]
```

```python
# plot the bar chart for revenue /course
regs.merge(course,on='course_id').groupby('course_name')['price'].sum().plot(kind='bar')
```

```python
#find the studnet who enrolled in both the months
commonStudent=np.intersect1d(nov['student_id'],dec['student_id'])
commonStudent
```

```python
students[students['student_id'].isin(commonStudent)]
```

```python
#find the student who no enrolled in any courses
students[~students['student_id'].isin(regs['student_id'])]
```

```python
#find the courses that go no enrolllment
#course['course_id]
#regs['course_id]

course_id_list=np.setdiff1d(course['course_id'],regs['course_id'])
course[course['course_id'].isin(course_id_list)]
```

#Self Join --> same table ko join kr lete hai


```python
#print student who did not enrolled into any courses
students.merge(students,how='inner',left_on='partner',right_on='student_id',suffixes=('_student','_partner'))
```

```python
#find the 3 students who did most number of enroolments
#regs.merge(students,on='student_id')['name'].value_counts()[:3] this not a suitable approach

regs.merge(students,on='student_id').groupby('name')['student_id'].count().sort_values(ascending=False)[:5]
#
```

```python
#10. find the top 3 students who spent most amount of money on courses
regs.merge(students, on='student_id') \
    .merge(course, on='course_id') \
    .groupby(['student_id', 'name'])['price'] \
    .sum() \
    .sort_values(ascending=False) \
    .head(3)
    #
```

```python
## alternate syntax for merge
#students.merge(regs)

pd.merge(students,regs,how='inner',on='student_id')
```
