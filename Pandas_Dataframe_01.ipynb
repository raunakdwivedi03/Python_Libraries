```python
#apply using in dataframe method
points_df = pd.DataFrame({
    '1st point': [(3,4), (-6,5), (0,0), (-10,1), (4,5)],
    '2nd point': [(-3,4), (0,0), (2,2), (10,10), (1,1)]
})

points_df
```

```python
def euclidian(row):
     x1,y1=row['1st point']
     x2,y2=row['2nd point']
     return np.sqrt((x2 - x1)**2 + (y2 - y1)**2)
```

```python
points_df['distance'] = points_df.apply(euclidian, axis=1)
points_df
```
