Pandas Series


column in excel and 1d array is called the series in pandas


```python
import pandas as pd
import numpy as np
```

Series from a lists(array)


```python
#string
country=['india','pakistan','usa','nepal','srilanka']

pd.Series(country) ## value and having index it is by default index aur uska value
                    ## dtype is object means pandas treat string as a object
```

```python
#integer
runs=[13,24,56,64,64]
pd.Series(runs)
```

```python
#custom index khud ka index bna skte hai
marks=[10,24,32,41,57,39]
subject=['maths','english','hindi','science','SSt','geography']

pd.Series(marks,index=subject)
```

```python
##setting a name to series
marks = pd.Series(marks,index=subject,name='raunak ke marks ')
marks
```

#Series from Dict

```python
marks={
    'maaths':67,
    'hindi':43,
    'english':10,
    'SSt':98
}

marks_series=pd.Series(marks,name='raunak ke marks ')
marks_series
```
